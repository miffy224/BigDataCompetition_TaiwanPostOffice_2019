```python
#ACC1:交寄局號
#ACC4:段號
#ACC10:責任局號
#ACC33:本頁重量(公斤)(該頁碼總重量)
```

```python
df=pd.read_sql('SELECT ACC1,ACC4,ACC10, ACC33 FROM ACC ', con=db_connection)
```

```python
#取每頁第一段
mask = df['ACC4'] == '1'
df=df[mask]
```

```python
data=df.sort_values(by = 'ACC1',ascending=True).reset_index().drop(['index', 'ACC4'], axis=1).rename(columns={'ACC1': 'O', 'ACC10': 'D', 'ACC33': 'weight(kg)'})
```

```python
result=data.groupby(['O', 'D']).sum()
```
