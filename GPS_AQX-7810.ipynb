```python
Post_box=pd.read_sql('SELECT x座標,y座標 FROM Post_box WHERE 縣市 IN ("臺南市")', con=db_connection)
Post_office=pd.read_sql('SELECT 經度,緯度 FROM Post_office WHERE 縣市 IN ("臺南市")', con=db_connection)
```

```python
office_seq=[]
box_seq=[]

for i in range(0,Post_office.index.size):
    office_seq.append([round(float(Post_office.緯度[i]),4), round(float(Post_office.經度[i]),4)])
for i in range(0,Post_box.index.size):
    box_seq.append([round(float(Post_box.y座標[i]),4), round(float(Post_box.x座標[i]),4)])
```

```python
from pandas.core.frame import DataFrame
o={"office" : office_seq}#将列表a成字典
office=DataFrame(o)#将字典转换成为数据框
print(office)


b={"box" : box_seq}#将列表a成字典
box=DataFrame(b)#将字典转换成为数据框
print(box)
```
